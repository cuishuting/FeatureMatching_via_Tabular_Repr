# This notebook transform those only one-time recorded itemids in mimic_chart_mv_only_num to the mimic_static_mv 

```python
import pandas as pd
```

```python
fill_limit_chart_mv = pd.read_csv("./data/fill_limit_chart_MV.csv")
item_id_with_no_remeasurement = fill_limit_chart_mv[fill_limit_chart_mv["recommend_fill_limit_MV"] == 0]["itemid"].values
mimic_chart_MV_only_num = pd.read_csv("../Processed_Data/mimic_chart_MV_only_numeric_new.csv")
mimic_chart_MV_only_num_one_time_itemid = mimic_chart_MV_only_num[mimic_chart_MV_only_num["itemid"].isin(item_id_with_no_remeasurement)]
mimic_chart_MV_only_num_one_time_itemid_part = mimic_chart_MV_only_num_one_time_itemid[["hadm_id", "itemid", "value"]]
added_static_mimic_mv = mimic_chart_MV_only_num_one_time_itemid_part.pivot(index='hadm_id', columns='itemid', values='value')
added_static_mimic_mv["hadm_id"] = added_static_mimic_mv.index
added_static_mimic_mv.reset_index(inplace=True, drop=True)
added_static_mimic_mv.head()
```

```python
mimic_static_MV = pd.read_csv("../Processed_Data/mimic_static_MV.csv")
mimic_static_MV
```

```python
mimic_static_MV_new = pd.merge(mimic_static_MV, added_static_mimic_mv, on="hadm_id", how="outer")
mimic_static_MV_new
```

```python
mimic_static_MV_new.to_csv("../Processed_Data/mimic_static_MV_added_onetime_itemid.csv", index=False)
```

# After this, don't forget to discard those one-time-recorded itemids' recording from "../Processed_Data/mimic_chart_MV_only_numeric_new.csv"

```python
cur_mimic_chart_mv_only_num = pd.read_csv("../Processed_Data/mimic_chart_MV_only_numeric_new.csv")
```

```python
print(cur_mimic_chart_mv_only_num.shape)
```

```python
one_time_recorded_itemid_chart_mv = mimic_chart_MV_only_num_one_time_itemid_part["itemid"].unique()
one_time_recorded_itemid_chart_mv
```

```python
cur_itemids_mimic_chart_mv_only_num = set(cur_mimic_chart_mv_only_num["itemid"].unique())
```

```python
print("before num of itemids:")
print(len(cur_itemids_mimic_chart_mv_only_num))
new_itemid_mimic_chart_mv_only_num = cur_itemids_mimic_chart_mv_only_num - set(one_time_recorded_itemid_chart_mv)
print("after num of itemids:")
print(len(new_itemid_mimic_chart_mv_only_num))
```

```python
new_mimic_chart_mv_only_num = cur_mimic_chart_mv_only_num[cur_mimic_chart_mv_only_num["itemid"].isin(new_itemid_mimic_chart_mv_only_num)]
new_mimic_chart_mv_only_num
```

```python
new_mimic_chart_mv_only_num.reset_index(inplace=True, drop=True)
new_mimic_chart_mv_only_num
```

```python
new_mimic_chart_mv_only_num.to_csv("../Processed_Data/mimic_chart_MV_only_num_del_onetime_itemid.csv", index=False)
```
