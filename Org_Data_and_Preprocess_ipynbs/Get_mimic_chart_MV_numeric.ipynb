```python
item_id_dbsource = pd.read_csv('./data/d_items_chartevents.csv')
```

```python
whole_mimic_chart_MV_reg_pre = pd.read_csv("./data/mimic_chart_MV_reg_pre.csv")

```

```python
print("original shape for whole_mimic_chart_MV_reg_pre: ", whole_mimic_chart_MV_reg_pre.shape)
mimic_chart_MV_numeric = whole_mimic_chart_MV_reg_pre[whole_mimic_chart_MV_reg_pre["itemid"].isin(list_itemid_num_chart_MV)]
print("after filtering only the numeric itemid, new shape is: ", mimic_chart_MV_numeric.shape)

```

```python
mimic_chart_MV_numeric.to_csv("./data/mimic_chart_MV_only_numeric.csv")
```
